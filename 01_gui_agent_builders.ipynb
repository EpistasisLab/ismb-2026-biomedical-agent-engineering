# GUI-Based Agent Building Platforms

This notebook introduces three popular GUI-based platforms for building AI agents — no deep coding required to get started. Each section covers installation, core concepts, and how to build your first agent.

| Platform | Best For | Hosting | Code Required |
|---|---|---|---|
| **N8N** | Workflow automation + API integration | Cloud or self-hosted | No |
| **Flowise AI** | LLM chatflows and RAG pipelines | Local or cloud | No |

## Part 1 — N8N

### What Is N8N?

[N8N](https://n8n.io) is an open-source workflow automation platform. You connect nodes visually to build pipelines that can call APIs, run AI agents, process files, send emails, and more. It has a built-in **AI Agent node** backed by any LLM you configure.

---

### Installation Options

**Option A — N8N Cloud (easiest, no setup)**
1. Go to [https://n8n.io](https://n8n.io) and sign up for a free trial.
2. Your workspace is ready immediately — no installation needed.

**Option B — Self-hosted via Docker (recommended for local use)**

```python
# Run in your terminal (not inside this notebook)
# Requires Docker Desktop: https://www.docker.com/products/docker-desktop/

setup_n8n = """
# Pull and start N8N with persistent storage
docker run -it --rm \\
  --name n8n \\
  -p 5678:5678 \\
  -v ~/.n8n:/home/node/.n8n \\
  n8nio/n8n

# Then open: http://localhost:5678
"""

print(setup_n8n)
```

**Option C — Self-hosted via npm**

```python
setup_n8n_npm = """
# Requires Node.js >= 18
npm install -g n8n
n8n start

# Then open: http://localhost:5678
"""

print(setup_n8n_npm)
```

---

### Core Concepts

| Concept | Description |
|---|---|
| **Workflow** | A canvas of connected nodes that execute in sequence or parallel |
| **Trigger node** | Starts the workflow (webhook, schedule, manual, chat) |
| **AI Agent node** | An LLM that reasons and calls tools |
| **Tool node** | A capability attached to the agent (web search, calculator, HTTP request) |
| **Memory node** | Gives the agent conversational memory across turns |
| **Credentials** | Stored API keys (OpenAI, Anthropic, Google, etc.) |

---

### Building Your First AI Agent in N8N

1. **Open** `http://localhost:5678` and create a new workflow.
2. **Add a Trigger**: drag in a *Chat Trigger* node (lets you chat with the agent in a sidebar).
3. **Add an AI Agent node**: connect it to the trigger.
   - Under *Chat Model*, select your LLM (e.g., Anthropic Claude or OpenAI GPT-4o).
   - Add your API key under *Credentials*.
4. **Add Tools** (optional): attach a *Wikipedia* or *SerpAPI* node as a tool to give the agent web access.
5. **Add Memory** (optional): attach a *Window Buffer Memory* node so the agent remembers the conversation.
6. **Activate** the workflow and click *Chat* in the top-right to test it.

### Tips
- Use the **HTTP Request** node to call any REST API as a custom tool.
- Use **Code** nodes (JavaScript or Python) to transform data between steps.
- Export workflows as JSON for version control (`⋮ → Download`).
- The [N8N community template library](https://n8n.io/workflows/) has hundreds of pre-built agent workflows.

## Part 2 — Flowise AI

### What Is Flowise?

[Flowise](https://flowiseai.com) is an open-source drag-and-drop UI for building LLM applications. It is purpose-built for LLM pipelines: RAG (retrieval-augmented generation), chatbots, and multi-agent flows. Built on top of LangChain and LlamaIndex under the hood.

---

### Installation

**Option A — npm (quickest)**

```python
setup_flowise_npm = """
# Requires Node.js >= 18
npm install -g flowise
npx flowise start

# With a username/password for local auth:
npx flowise start --FLOWISE_USERNAME=admin --FLOWISE_PASSWORD=yourpassword

# Then open: http://localhost:3000
"""

print(setup_flowise_npm)
```

**Option B — Docker**

```python
setup_flowise_docker = """
docker run -d \\
  --name flowise \\
  -p 3000:3000 \\
  -v ~/.flowise:/root/.flowise \\
  flowiseai/flowise

# Then open: http://localhost:3000
"""

print(setup_flowise_docker)
```

---

### Core Concepts

| Concept | Description |
|---|---|
| **Chatflow** | A pipeline for conversational LLM apps (Q&A, RAG) |
| **Agentflow** | A multi-step agentic pipeline with branching and tool use |
| **Node** | A component — LLM, vector store, tool, memory, etc. |
| **Document Loader** | Ingests files (PDF, CSV, web pages) into a vector store |
| **Vector Store** | Stores embeddings for semantic retrieval (e.g., Chroma, Pinecone) |
| **Credentials** | Stored API keys managed in the UI |

---

### Building a RAG Chatbot in Flowise

1. **Open** `http://localhost:3000` → click **Add New** under *Chatflows*.
2. **Add a Document Loader**: drag in *PDF File Loader* and upload a document.
3. **Add a Text Splitter**: connect a *Recursive Character Text Splitter* to chunk the document.
4. **Add a Vector Store**: connect an *In-Memory Vector Store* (no setup needed) or *Chroma* for persistence.
5. **Add an Embeddings node**: connect *OpenAI Embeddings* or *Ollama Embeddings* to the vector store.
6. **Add a Conversational Retrieval QA Chain**: this wires together the retriever + LLM + memory.
7. **Add a Chat Model**: connect Claude or GPT-4o; add your API key under *Credentials*.
8. **Save and test** using the chat bubble in the bottom-right corner.

### Building a Multi-Agent Flow in Flowise

1. Go to **Agentflows** → **Add New**.
2. Add a **Supervisor** agent node — this orchestrates the team.
3. Add **Worker** agent nodes (e.g., a Researcher and a Writer), each with their own tools.
4. Connect Workers to the Supervisor.
5. Attach **Tools** (Calculator, Web Browser, Custom Tool) to individual workers.
6. Test with the built-in chat interface.

### Tips
- Use **Ollama** nodes to run local models (Llama 3, Mistral) completely offline.
- Flows can be exported as JSON and version-controlled.
- Flowise exposes a REST API for each chatflow — share or embed it in any app.
- The [Flowise template marketplace](https://flowiseai.com/templates) has ready-made RAG and agent flows.

## Part 3 — AutoGen Studio

### What Is AutoGen Studio?

[AutoGen Studio](https://microsoft.github.io/autogen/docs/autogen-studio/getting-started) is a web UI built on Microsoft's [AutoGen](https://github.com/microsoft/autogen) framework. It lets you design, test, and iterate on **multi-agent teams** without writing orchestration code. Agents can collaborate, debate, write code, execute it, and hand off tasks to each other.

---

### Hosting a Local Instance

**Step 1 — Create and activate a Python environment**

```python
setup_env = """
# Using conda (recommended)
conda create -n autogen python=3.11 -y
conda activate autogen

# Or using venv
python -m venv autogen_env
source autogen_env/bin/activate   # macOS/Linux
autogen_env\\Scripts\\activate      # Windows
"""

print(setup_env)
```

**Step 2 — Install AutoGen Studio**

```python
# Install directly via pip
import subprocess
result = subprocess.run(
    ["pip", "install", "autogenstudio"],
    capture_output=True, text=True
)
print(result.stdout)
print(result.stderr)
```

**Step 3 — Set your LLM API key**

```python
import os

# Set whichever key matches your chosen model
os.environ["OPENAI_API_KEY"] = "sk-..."       # OpenAI
# os.environ["ANTHROPIC_API_KEY"] = "sk-ant-..."  # Anthropic Claude

# For persistent use, add the export line to your shell profile:
# export OPENAI_API_KEY="sk-..."
print("API key set.")
```

**Step 4 — Launch the local server**

```python
launch_cmd = """
# Run in your terminal (or uncomment the subprocess call below to start from this notebook)
autogenstudio ui --port 8081 --host 0.0.0.0

# Then open: http://localhost:8081

# Optional flags:
#   --appdir ./my_data    store sessions/agents in a custom directory
#   --reload              auto-reload server on code changes (dev mode)
"""
print(launch_cmd)

# To start the server programmatically from this notebook:
# import subprocess, sys
# proc = subprocess.Popen([sys.executable, "-m", "autogenstudio", "ui", "--port", "8081"])
# print(f"AutoGen Studio started (PID {proc.pid}) → http://localhost:8081")
```

---

### Navigating the AutoGen Studio UI

Once you open `http://localhost:8081`, you will see four main sections:

| Section | Purpose |
|---|---|
| **Playground** | Chat with your agent teams and inspect message traces |
| **Build → Agents** | Define individual agents (name, system prompt, tools, model) |
| **Build → Teams** | Assemble agents into a team with an orchestration strategy |
| **Build → Models** | Register LLM configurations (model name, API key, base URL) |

---

### Building Your First Multi-Agent Team

**Step 1 — Register a model**
1. Go to **Build → Models → + New Model**.
2. Set *Model* to `gpt-4o` (or `claude-3-5-sonnet-20241022` for Anthropic).
3. Paste your API key and click **Save**.

**Step 2 — Create agents**
1. Go to **Build → Agents → + New Agent**.
2. Create a **Researcher** agent:
   - *System Message*: `You are a research assistant. Find accurate, concise information on the topic given to you.`
   - Attach the model you registered.
3. Create a **Writer** agent:
   - *System Message*: `You are a scientific writer. Take research notes and write a clear, structured summary.`
   - Attach the same or a different model.
4. Add a **UserProxyAgent** — this represents the human in the loop and can run code locally if needed.

**Step 3 — Assemble a team**
1. Go to **Build → Teams → + New Team**.
2. Choose an orchestration strategy:
   - *RoundRobin*: agents take turns in order.
   - *Selector* (recommended): a Selector LLM picks which agent speaks next based on context.
   - *Swarm*: agents hand off autonomously using tool calls.
3. Add your Researcher, Writer, and UserProxyAgent to the team.
4. Set a **termination condition** (e.g., stop after 10 messages, or when the string `TERMINATE` appears).
5. Click **Save**.

**Step 4 — Test in the Playground**
1. Go to **Playground → + New Session**.
2. Select your team.
3. Type a task (e.g., `Summarize the key applications of CRISPR in cancer therapy.`).
4. Watch the agents collaborate — click any message to inspect the full agent trace.

---

### Enabling Code Execution

AutoGen agents can write and execute Python code locally. To enable this safely:

```python
code_execution_note = """
In AutoGen Studio, when creating a UserProxyAgent:
  - Set 'Code Execution' to ENABLED
  - Set 'Work Directory' to a sandboxed folder, e.g. ./autogen_workspace
  - Optionally enable Docker-based execution for isolation:

    code_execution_config = {
        "use_docker": True,          # runs code in a container
        "work_dir": "./workspace"
    }

WARNING: only enable local code execution in a trusted environment.
Use Docker isolation if running untrusted agent-generated code.
"""
print(code_execution_note)
```

---

### Tips
- **Export teams** as JSON from the UI for sharing and version control.
- **System messages matter** — clear role instructions dramatically improve agent collaboration.
- **Max messages**: always set a termination condition to prevent runaway agent loops.
- **Ollama integration**: register a local Ollama model by setting the *Base URL* to `http://localhost:11434/v1` and model to e.g. `llama3.2`.
- After prototyping in the UI, export to Python code using the AutoGen SDK for production use.
